```python
df['committers2'] = df['committers'].apply(lambda row: json.loads(row.replace("'", '"').replace('False', '"False"').replace('True', '"True"')))
```

```python
to_hist = df['committers_count']
```

```python
# for spine in plt.gca().spines.values():
#     spine.set_visible(False)

plt.figure(figsize=[10,8])
n, bins, patches = plt.hist(x=to_hist, bins=25, color='#5b2084',alpha=0.7, rwidth=0.85)
# plt.grid(axis='y', alpha=0.75)
plt.xlabel('Commiters count',fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
# plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')

plt.tick_params(
    right=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)


plt.show()
```
